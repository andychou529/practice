```python
import matplotlib.pyplot as plt

listx = ["","","","","","","",""]
listm = [0.14,0.16,0.08,0.13,0.16,0.12,0.16,0.05]
listf = [0.1,0.19,0.06,0.1,0.13,0.13,0.2,0.09]

listm = [x*100 for x in listm]
listf = [x*100 for x in listf]

plt.figure(figsize=[12,9])
plt.rcParams['font.sans-serif'] = 'DFKai-SB '
plt.rcParams['axes.unicode_minus'] = False

plt.subplot(221)
plt.title(':',fontsize=16)
plt.pie(listm,labels=listx,autopct='%2.1f%%')

plt.subplot(222)
plt.title(':',fontsize=16)
plt.pie(listf,labels=listx,autopct='%2.1f%%')

plt.subplot(223)
width = 0.4
listx1=[x -  width/2 for x in range (len(listx))]
listx2=[x +  width/2 for x in range (len(listx))]

plt.title('-',fontsize=16)
plt.xlabel('',fontsize=12)
plt.ylabel('(%)',fontsize=12)

plt.bar(listx1,listm,width,label='')
plt.bar(listx2,listf,width,label='')
plt.xticks(range(len(listx)),listx,rotation=45)
plt.legend()


plt.subplot(224)
plt.title('-',fontsize=16)
plt.xlabel('',fontsize=12)
plt.ylabel('(%)',fontsize=12)

plt.plot(listx,listm,marker="s",label='')
plt.plot(listx,listf,marker="s",label='')
plt.gca().grid(True)
plt.xticks(rotation=45)
plt.legend()

plt.show()
```
